```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_titan = pd.read_csv("Titanic-Dataset.csv")
```

```python
df_titan.head(10)
```

```python
df_inval = df_titan[df_titan['Age'].notnull()]
```

```python
df_inval
```

```python
df_t = df_inval[['Name','Sex','Age','Fare']].head(40)
```

```python
df_t['category'] = df_t['Age']
```

```python
def category(Age):
    if Age <= 1:
        return 'Bebe'
    elif Age >1 and Age <=12:
        return 'Niño'
    elif Age >12 and Age <= 18:
        return 'Adolecente'
    elif Age > 18 and Age <= 55:
        return 'Adulto'
    else:
        return 'Adulto mayor'
```

```python
df_t['category'] = df_t['Age'].apply(category)
```

```python
df_t
```

```python
#Nombres con edades similares
df_t.groupby(['Age', 'Name']).count()
```

```python
df_t.groupby(['Name', 'category'], sort = False).count()
```
